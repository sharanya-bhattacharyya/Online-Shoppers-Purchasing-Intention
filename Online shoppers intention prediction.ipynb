```python
description = pd.DataFrame(index=['observations(rows)', 'percent missing', 'dtype', 'range'])
numerical = []
categorical = []


for col in df4.columns:
    obs = df4[col].size
    p_nan = round(df4[col].isna().sum()/obs, 2) # caluclating the number of missing values
    num_nan = f'{p_nan}% ({df4[col].isna().sum()}/{obs})' # caluclating the number of missing values
    dtype = 'categorical' if df4[col].dtype == object else 'numerical' # defining the criteria for numerical and categorical columns
    numerical.append(col) if dtype == 'numerical' else categorical.append(col) # preparing the list of categorcal and numerical columns
    rng = f'{len(df4[col].unique())} labels' if dtype == 'categorical' else f'{df4[col].min()}-{df4[col].max()}' # the range of values in the columns
    description[col] = [obs, num_nan, dtype, rng]
```

## Model Building

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report
from sklearn import metrics
classifier = RandomForestClassifier(n_estimators=20)
classifier.fit(X_train,y_train)
classifier.score(X_train,y_train)
```

```python
classifier.score(X_test,y_test)
```

```python
n_estimators = [100, 300, 500, 800, 1200]
max_depth = [5, 8, 15, 25, 30]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10] 

param_grid = dict(n_estimators = n_estimators, max_depth = max_depth,  
              min_samples_split = min_samples_split, 
             min_samples_leaf = min_samples_leaf)
classifier_cv = RandomizedSearchCV(classifier,param_grid,cv=5)
```

```python
classifier_cv.fit(X_train,y_train)
classifier_cv.best_params_
```

```python
classifier1 = RandomForestClassifier(n_estimators=500,min_samples_split=2,min_samples_leaf=2,max_depth=15)
classifier1.fit(X_train,y_train)
y_pred = classifier1.predict(X_test)
report=classification_report(y_test, y_pred)
print(report)
```

```python
classifier1.score(X_train, y_train)
```

```python
print(metrics.confusion_matrix(y_test, y_pred))
```

```python
classifier1.score(X_test,y_test)
```

```python
pscore = metrics.accuracy_score(y_test, y_pred)
pscore
```

### Turning Imbalanced Dataset to a Balanced Dataset

```python
from imblearn.over_sampling import SMOTE
smote= SMOTE('minority')
X_sm,y_sm = smote.fit_resample(X_train,y_train)
print(X_sm.shape,y_sm.shape)
```

```python
y_sm.Returning_Visitor.value_counts()/y_sm.Returning_Visitor.shape[0]
```

```python
classifier_cv.fit(X_sm,y_sm)
classifier_cv.best_params_
```

```python
classifier1 = RandomForestClassifier(n_estimators=300,min_samples_split=10,min_samples_leaf=2,max_depth=30)
classifier1.fit(X_sm,y_sm)
y_pred = classifier1.predict(X_test)
report=classification_report(y_test, y_pred)
print(report)
```

```python
classifier1.score(X_sm,y_sm)
```

```python
classifier1.score(X_test,y_test)
```

```python
print(metrics.confusion_matrix(y_test, y_pred))
```

```python
pscore = metrics.accuracy_score(y_test, y_pred)
pscore
```

### Checking feature Importance based on model

```python
for feature in zip(X_sm.columns, classifier1.feature_importances_):
    print(feature)
```

```python
from sklearn.feature_selection import SelectFromModel
sel = SelectFromModel(RandomForestClassifier(n_estimators=300,min_samples_split=10,min_samples_leaf=1,max_depth=25))
sel.fit(X_sm, y_sm)
```

```python
sel.get_support()
```

```python
selected_feat= X_train.columns[(sel.get_support())]
len(selected_feat)
```

```python
print(selected_feat)
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 20), dpi=80, facecolor='w', edgecolor='k')
features = X_sm.columns
importances = classifier1.feature_importances_
indices = np.argsort(importances)
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

ns_probs = [0 for _ in range(len(y_test))]

# predict probabilities
lr_probs = classifier1.predict_proba(X_test)
# keep probabilities for the positive outcome only
lr_probs = lr_probs[:, 1]
# calculate scores
ns_auc = roc_auc_score(y_test, ns_probs)
lr_auc = roc_auc_score(y_test, lr_probs)
# summarize scores
print('Returning Visitor: ROC AUC=%.3f' % (ns_auc))
print('New customer: ROC AUC=%.3f' % (lr_auc))
# calculate roc curves
ns_fpr, ns_tpr, _ = roc_curve(y_test, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
# plot the roc curve for the model
plt.plot(ns_fpr, ns_tpr, linestyle='--', label='Returning Visitor')
plt.plot(lr_fpr, lr_tpr, marker='.', label='New customer')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```
